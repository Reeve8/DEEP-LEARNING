```python
import matplotlib.pyplot as plt

# Create a figure and an axis
fig, ax = plt.subplots()

# Draw the apple (a red circle)
apple = plt.Circle((0.5, 0.5), 0.4, color='red', edgecolor='black')
ax.add_patch(apple)

# Draw the leaf (a green polygon)
leaf = plt.Polygon([[0.5, 0.8], [0.6, 1.0], [0.7, 0.8]], color='green')
ax.add_patch(leaf)

# Draw the stem (a brown line)
stem = plt.Line2D([0.5, 0.5], [0.7, 0.9], color='brown', linewidth=2)
ax.add_line(stem)

# Set the aspect ratio of the plot to be equal
ax.set_aspect('equal')

# Remove the axes
ax.axis('off')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure and an axis
fig, ax = plt.subplots()

# Draw the watermelon body (a green circle)
watermelon_body = plt.Circle((0.5, 0.5), 0.4, color='green', edgecolor='black')
ax.add_patch(watermelon_body)

# Draw the watermelon flesh (a red semi-circle)
theta = np.linspace(0, np.pi, 100)
x = 0.5 + 0.4 * np.cos(theta)
y = 0.5 + 0.4 * np.sin(theta)
ax.fill_between(x, y, 0.5, color='red')

# Draw watermelon seeds
seed_positions = [(0.5, 0.7), (0.6, 0.65), (0.4, 0.65), (0.55, 0.55), (0.45, 0.55)]
for (x, y) in seed_positions:
    seed = plt.Circle((x, y), 0.02, color='black')
    ax.add_patch(seed)

# Set the aspect ratio of the plot to be equal
ax.set_aspect('equal')

# Remove the axes
ax.axis('off')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

# Draw the banana (a yellow ellipse)
banana = plt.Ellipse((0.5, 0.5), 0.8, 0.4, angle=30, color='yellow', edgecolor='black')
ax.add_patch(banana)

# Set the aspect ratio of the plot to be equal
ax.set_aspect('equal')
ax.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Draw grapes (multiple purple circles)
for i in range(3):
    for j in range(3):
        grape = plt.Circle((0.5 + i*0.1, 0.5 - j*0.1), 0.05, color='purple', edgecolor='black')
        ax.add_patch(grape)

# Set the aspect ratio of the plot to be equal
ax.set_aspect('equal')
ax.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Draw the orange (an orange circle)
orange = plt.Circle((0.5, 0.5), 0.4, color='orange', edgecolor='black')
ax.add_patch(orange)

# Draw the segments (white lines)
for i in range(12):
    angle = np.pi / 6 * i
    x = 0.5 + 0.4 * np.cos(angle)
    y = 0.5 + 0.4 * np.sin(angle)
    line = plt.Line2D([0.5, x], [0.5, y], color='white', linewidth=2)
    ax.add_line(line)

# Set the aspect ratio of the plot to be equal
ax.set_aspect('equal')
ax.axis('off')

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

# Draw the strawberry body (a red heart shape)
theta = np.linspace(0, 2*np.pi, 100)
x = 0.5 + 0.3 * np.sin(theta) * np.cos(theta)
y = 0.5 + 0.3 * np.sin(theta)
strawberry = plt.Polygon(np.column_stack([x, y]), color='red', edgecolor='black')
ax.add_patch(strawberry)

# Draw the strawberry leaves (green triangles)
for i in range(5):
    angle = 2 * np.pi / 5 * i
    leaf_x = [0.5 + 0.1 * np.cos(angle), 0.5 + 0.15 * np.cos(angle + np.pi / 5), 0.5 + 0.1 * np.cos(angle + 2 * np.pi / 5)]
    leaf_y = [0.5 + 0.3 + 0.1 * np.sin(angle), 0.5 + 0.3 + 0.15 * np.sin(angle + np.pi / 5), 0.5 + 0.3 + 0.1 * np.sin(angle + 2 * np.pi / 5)]
    leaf = plt.Polygon(np.column_stack([leaf_x, leaf_y]), color='green', edgecolor='black')
    ax.add_patch(leaf)

# Set the aspect ratio of the plot to be equal
ax.set_aspect('equal')
ax.axis('off')

plt.show()
```
